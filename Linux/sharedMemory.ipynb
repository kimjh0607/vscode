네, CentOS Linux에서 공유 메모리를 설정하는 데 있어 `tmpfs`를 사용하여 `/dev/shm` 디렉토리의 크기를 조정하는 방법도 있습니다. `tmpfs`는 메모리 기반 파일 시스템으로, 메모리를 가상 파일 시스템으로 마운트하여 사용할 수 있게 합니다.

### `tmpfs`를 사용한 공유 메모리 설정 조정

1. **현재 `/dev/shm` 크기 확인**

   현재 `/dev/shm`의 크기를 확인하려면 다음 명령어를 사용합니다:

   ```sh
   df -h /dev/shm
   ```

2. **`/etc/fstab` 파일 편집**

   `vi` 또는 다른 텍스트 편집기를 사용하여 `/etc/fstab` 파일을 엽니다:

   ```sh
   sudo vi /etc/fstab
   ```

3. **`tmpfs` 설정 추가 또는 수정**

   `/etc/fstab` 파일에 다음 줄을 추가하거나 수정하여 `/dev/shm`의 크기를 설정합니다:

   ```sh
   tmpfs   /dev/shm   tmpfs   defaults,size=4G   0  0
   ```

   여기서 `size=4G`는 `/dev/shm`의 크기를 4GB로 설정하는 예입니다. 필요한 크기로 조정할 수 있습니다.

4. **마운트 재설정**

   설정을 적용하려면 `/dev/shm`을 다시 마운트합니다:

   ```sh
   sudo mount -o remount /dev/shm
   ```

5. **확인**

   설정이 적용되었는지 확인하려면 다시 한 번 `/dev/shm`의 크기를 확인합니다:

   ```sh
   df -h /dev/shm
   ```

### 추가적인 공유 메모리 설정

앞서 언급한 `kernel.shmmax`와 `kernel.shmall` 외에도 공유 메모리와 관련된 다른 커널 파라미터가 있습니다:

- **`kernel.shmmni`**: 시스템에서 허용되는 최대 공유 메모리 세그먼트 수.
- **`kernel.sem`**: 세마포어 설정으로, 공유 메모리와 관련된 다른 IPC 리소스를 조정합니다.

이 파라미터들은 `/etc/sysctl.conf` 파일에서 설정할 수 있으며, 설정 후 `sysctl -p` 명령어를 사용하여 적용할 수 있습니다.

예를 들어:

```sh
sudo vi /etc/sysctl.conf
```

```sh
kernel.shmmax = 68719476736   # 64GB
kernel.shmall = 16777216      # 16M 페이지 (64GB)
kernel.shmmni = 4096          # 최대 4096개의 공유 메모리 세그먼트
kernel.sem = 250 32000 100 128 # 세마포어 설정
```

설정 후 적용:

```sh
sudo sysctl -p
```

이러한 설정을 통해 CentOS에서 공유 메모리와 관련된 다양한 파라미터를 조정할 수 있습니다.
