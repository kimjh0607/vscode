## 분석 및 설명


### listener.ora 파일

`listener.ora` 파일은 Oracle Database의 리스너를 구성하는 파일입니다. 이 파일은 데이터베이스 인스턴스가 클라이언트 요청을 수신할 수 있도록 설정합니다.

내용:
```sh
# listener.ora Network Configuration File: /app/oracle/product/11.2.0/dbhome_1/network/admin/listener.ora
# Generated by Oracle configuration tools.


SID_LIST_LISTENER =
   (SID_LIST =
     (SID_DESC =
       (SID_NAME = orcl)
       (ORACLE_HOME = /app/oracle/product/11.2.0/dbhome_1)
     )
     (SID_DESC =
       (SID_NAME = orcl2)
       (ORACLE_HOME = /app/oracle/product/11.2.0/dbhome_1)
     )
   )

LISTENER =
   (DESCRIPTION_LIST =
     (DESCRIPTION =
       (ADDRESS = (PROTOCOL = TCP)(HOST = 192.168.0.229)(PORT = 1521))
       (ADDRESS = (PROTOCOL = IPC)(KEY = EXTPROC1521))
     )
   )

LISTENER2 =
   (DESCRIPTION_LIST =
     (DESCRIPTION =
       (ADDRESS = (PROTOCOL = TCP)(HOST = 192.168.0.229)(PORT = 1522))
       (ADDRESS = (PROTOCOL = IPC)(KEY = EXTPROC1522))
     )
   )

ADR_BASE_LISTENER = /app/oracle
```

#### 설명:
- `SID_LIST_LISTENER`는 리스너에 연결된 데이터베이스 인스턴스를 정의합니다. 여기에는 두 개의 인스턴스 `orcl`과 `orcl2`가 있습니다.
- `LISTENER`와 `LISTENER2`는 각각 다른 포트에서 작동하는 두 개의 리스너를 정의합니다. `LISTENER`는 포트 1521에서, `LISTENER2`는 포트 1522에서 작동합니다.
- `ADDRESS`는 TCP와 IPC 프로토콜을 사용하여 호스트와 포트를 지정합니다.
- `ADR_BASE_LISTENER`는 ADR(Automatic Diagnostic Repository) 기반 위치를 지정합니다.

### tnsnames.ora 파일

`tnsnames.ora` 파일은 클라이언트가 데이터베이스에 연결하기 위해 사용하는 구성 파일입니다.

내용:
```sh
# tnsnames.ora Network Configuration File: /app/oracle/product/11.2.0/dbhome_1/network/admin/tnsnames.ora
# Generated by Oracle configuration tools.

LISTENER_ORCL2 =
  (ADDRESS = (PROTOCOL = TCP)(HOST = 192.168.0.229)(PORT = 1522))


ORACLR_CONNECTION_DATA =
  (DESCRIPTION =
    (ADDRESS_LIST =
      (ADDRESS = (PROTOCOL = IPC)(KEY = EXTPROC1521))
    )
   (CONNECT_DATA =
     (SID = CLRExtProc)
      (PRESENTATION = RO)
    )
  )

LISTENER_ORCL =
  (ADDRESS = (PROTOCOL = TCP)(HOST = 192.168.0.229)(PORT = 1521))


ORCL2 =
  (DESCRIPTION =
    (ADDRESS = (PROTOCOL = TCP)(HOST = 192.168.0.229)(PORT = 1522))
    (CONNECT_DATA =
      (SERVER = DEDICATED)
      (SERVICE_NAME = orcl2)
    )
  )

ORCL =
  (DESCRIPTION =
    (ADDRESS = (PROTOCOL = TCP)(HOST = 192.168.0.229)(PORT = 1521))
    (CONNECT_DATA =
      (SERVER = DEDICATED)
      (SERVICE_NAME = orcl)
    )
  )
```

#### 설명:
- `LISTENER_ORCL2`와 `LISTENER_ORCL`는 각각 포트 1522와 1521에서 작동하는 리스너를 지정합니다.
- `ORCL2`와 `ORCL`은 각각 `orcl2`와 `orcl` 서비스 이름으로 데이터베이스에 연결하기 위한 설정입니다. 두 항목 모두 TCP 프로토콜을 사용하여 `192.168.0.229` 호스트의 해당 포트로 연결합니다.
- `ORACLR_CONNECTION_DATA`는 IPC 프로토콜을 사용하여 `EXTPROC1521` 키를 통해 연결하는 설정입니다.

### 항목의 차이점
- `LISTENER_ORCL`: 리스너의 네트워크 주소를 정의합니다. 이는 주로 리스너에 연결하는 데 사용됩니다.
- `ORCL`: 데이터베이스 서비스에 연결하는 데 필요한 전체 정보를 정의합니다. 이는 클라이언트가 데이터베이스 인스턴스에 연결하는 데 사용됩니다.

### 논리적 연계

1. **리스너 구성 (`listener.ora`)**:
   - 리스너는 클라이언트 연결 요청을 수신하고 데이터베이스 인스턴스에 전달하는 역할을 합니다.
   - 두 개의 리스너(`LISTENER`, `LISTENER2`)가 서로 다른 포트에서 작동하여 다중 인스턴스를 지원합니다.
   
2. **클라이언트 구성 (`tnsnames.ora`)**:
   - 클라이언트는 `tnsnames.ora` 파일을 참조하여 올바른 리스너와 포트를 통해 데이터베이스 인스턴스에 연결합니다.
   - 각 데이터베이스 인스턴스(`orcl`, `orcl2`)에 대한 연결 설정이 정의되어 있습니다.

### 결론

`listener.ora`와 `tnsnames.ora` 파일은 다중 인스턴스 환경에서 각각의 인스턴스가 고유의 리스너를 통해 클라이언트 요청을 수신할 수 있도록 구성됩니다. 이러한 구성을 통해 고가용성과 부하 분산을 제공할 수 있습니다.
